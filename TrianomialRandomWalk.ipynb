```python
import numpy as np
import matplotlib.pyplot as plt
import random
def trinomial_random_walk(steps, up_prob, down_prob, no_change_prob, up_factor, down_factor):
    n = 100  # Number of random numbers
    li = random.sample(range(1, 101), n)
    prices = li  # Start with an initial price of 1
    for _ in range(steps):
        rand = np.random.rand()
        if rand < up_prob:
            prices.append(prices[-1] * up_factor)
        elif rand < up_prob + down_prob:
            prices.append(prices[-1] * down_factor)
        else:
            prices.append(prices[-1])  # No change
    return prices

# Parameters
steps = 100
up_prob = 0.3
down_prob = 0.3
no_change_prob = 0.4
up_factor = 1.1
down_factor = 0.9
```

```python

# Simulate and plot
prices = trinomial_random_walk(steps, up_prob, down_prob, no_change_prob, up_factor, down_factor)
plt.plot(prices)
plt.title("Trinomial Random Walk")
plt.xlabel("Steps")
plt.ylabel("Price")
plt.show()
```
